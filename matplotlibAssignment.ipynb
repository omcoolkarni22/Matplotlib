```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**1.Read data from survey_results_public.csv in dataframe object**

```python
df = pd.read_csv('survey_results_public.csv')
df.head(5)
```

```python
ind = df[df['Country'] == 'India']
```

```python
ind
```

```python
ind.dropna(inplace=True)
```

```python
ind
```

**2. Plot Age column distribution(histogram) of developers from india only so we
get idea how many developers of each age group from india took part in servey**

```python
plt.title('India Devs Age')
plt.xlabel('Age')
plt.ylabel('Total')
arr = np.arange(10, 90, 10)
plt.hist(ind['Age'], bins=arr, edgecolor='black')
```

```python
df.dropna(inplace=True)
```

**3. Plot pie chart by using Gender column to show count of mail and female took
part in survey**

```python
gen = pd.Series(df['Gender'].value_counts())
```

```python
gen.values[:2]
```

```python
gen.keys()[:2]
```

```python
df['Gender'].value_counts().values
```

```python
plt.title('Gender Pie Chart')
plt.pie(gen.values[:2],labels=gen.keys()[:2], autopct="%.2f%%")
plt.legend(title='Gender', loc='center left', bbox_to_anchor=(1.2, 0.5))
plt.axis('equal')
```
