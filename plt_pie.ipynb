```python
help(plt.pie)
```

```python
values  = [251, 135, 452, 754]
lang = ['Python', 'Java', 'C++', 'C']
colors = ['red', 'green', 'blue', 'gold']
explode = [0.05, 0.05, 0.05, 0.05]
```

```python
plt.pie(values, labels=lang, colors=colors, shadow=True, autopct='%.1f%%', explode=explode)
#for equal axis, perfect pie
plt.legend(lang)
plt.axis('equal')
```
