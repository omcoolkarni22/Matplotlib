```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df
```

```python
x = df['gold_medal'].values
```

```python
labels = df['country'].values
```

```python
explode = [0.05, 0.05, 0.05, 0.05, 0.05]
```

```python
plt.pie(x, labels=labels, autopct='%.1f%%', explode=explode, shadow=True)
plt.legend(loc='best', title='Countries')
plt.axis('equal')
```
